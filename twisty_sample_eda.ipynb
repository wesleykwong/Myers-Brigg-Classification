# Bar Plot of Tweet Distribution

Number of samples for each language:

```python
twitter_counts = twisty_sample.groupby('language').count()[['mbti']].reset_index().rename(columns={'mbti':'counts'})
twitter_counts
```

MBTI Distribution (Raw Counts):

```python
mbti_distribution = twisty_sample.groupby(['mbti', 'language']).count()
mbti_distribution = mbti_distribution.drop(['user_id', 'twitter_text'], axis=1).rename(columns={'confirmed_tweet_ids':'count'})
mbti_distribution
```

```python
total_mbti_distribution = mbti_distribution.reset_index().groupby('mbti').sum()
total_mbti_distribution_percent = np.round(total_mbti_distribution/np.sum(total_mbti_distribution)*100, 3)
total_mbti_distribution_percent = total_mbti_distribution_percent.rename(columns={'count':'language'})

mbti_distribution_language_percent = mbti_distribution.reset_index()
```

```python
# Percentage of MBTI distribution over all samples
total_mbti_distribution_percent = total_mbti_distribution_percent.reset_index().rename(columns={'language':'count'})
total_mbti_distribution_percent['language'] = 'total'
total_mbti_distribution_percent
```

```python
def mbti_distribution_language(language):
    total_language_counts = int(twitter_counts[twitter_counts['language'] == language]['counts'])
    
    mbti_distribution_language = mbti_distribution_language_percent[mbti_distribution_language_percent['language'] == language]
    mbti_distribution_language = mbti_distribution_language.set_index('mbti').drop(columns=['language'])
    
    percentage = np.round(mbti_distribution_language/total_language_counts*100, 3)
    percentage['language'] = language
    
    return percentage.reset_index()
    
```

```python
mbti_distribution_percent = pd.concat([mbti_distribution_language('german'),
                                       mbti_distribution_language('spanish'),
                                       mbti_distribution_language('italian'),
                                       mbti_distribution_language('dutch'),
                                       total_mbti_distribution_percent], 
                                    axis=0)
```

```python
plot_mbti_distribution_percent = mbti_distribution_percent.set_index(['mbti', 'language']).reset_index()
plot_mbti_distribution_percent = plot_mbti_distribution_percent.rename(columns={'language':'Data'})
plot_mbti_distribution_percent
```

```python
# Created a list of the mbti classes in order based on the total percentage decreasing
order_mbti = plot_mbti_distribution_percent[plot_mbti_distribution_percent['Data'] == 'total'].sort_values(
    by='count', ascending=False)['mbti']
```

```python
# Cleans up the variable and tags for visualization
plot_mbti_distribution_percent['Data'] = plot_mbti_distribution_percent['Data'].str.title()
plot_mbti_distribution_percent = plot_mbti_distribution_percent.rename(
    columns={'mbti':'MBTI Class', 'count':'Percentage of MBTI Tweets'})
plot_mbti_distribution_percent
```

```python
sns.color_palette("hls", 16)
sns.catplot(x="Percentage of MBTI Tweets", y="MBTI Class", col="Data",
            order=order_mbti,
            col_order=['Total', 'German', 'Spanish', 'Italian', 'Dutch'],
            data=plot_mbti_distribution_percent, kind="bar");

# Better to drag it out as a image onto desktop since savefig cuts off the bottom
#plt.savefig('mbti_distribution_by_language.png')
```

# Bar Plot of Author Distribution

```python
author_count = twisty_sample.groupby(['language', 'user_id']).count()['mbti']
author_count = author_count.reset_index().groupby('language').count()['user_id'].reset_index()
author_count
```

```python
groupby_author = twisty_sample.groupby(['language', 'user_id', 'mbti']).count()['twitter_text'].reset_index()
groupby_author
```

```python
total_author = groupby_author.drop(columns=['twitter_text', 'language'])
total_author = total_author.groupby('mbti').count()
total_author['percentage'] = total_author['user_id']/total_author['user_id'].sum()*100
total_author = total_author.drop(columns=['user_id']).reset_index() 
total_author['language'] = 'total'
total_author
```

```python
def mbti_distribution_author(language):
    author = groupby_author[groupby_author['language'] == language].drop(columns=['twitter_text'])
    author = author.groupby('mbti').count()['language'].reset_index().copy()
    author['percentage'] = author['language']/author['language'].sum()*100
    author = author.drop(columns=['language'])
    author['language'] = language
    return author
```

```python
author_dist = pd.concat([mbti_distribution_author(x) for x in languages_data], axis=0)
author_dist = pd.concat([total_author, author_dist], axis=0)
author_dist 
```

```python
# Created a list of the mbti classes in order based on the total percentage decreasing
mbti_order_author = total_author.sort_values(by='percentage', ascending=False)['mbti']
```

```python
# Cleans up the variable and tags for visualization
author_dist ['language'] = author_dist ['language'].str.title()
author_dist  = author_dist .rename(columns={'mbti':'MBTI Class', 
                                            'percentage':'Percentage of Authors',
                                           'language': 'Data'})
author_dist 
```

```python
sns.color_palette("hls", 16)
sns.catplot(x="Percentage of Authors", y="MBTI Class", col="Data",
            order=mbti_order_author,
            col_order=['Total', 'German', 'Spanish', 'Italian', 'Dutch'],
            data=author_dist , kind="bar");

# Better to drag it out as a image onto desktop since savefig cuts off the bottom
#plt.savefig('mbti_distribution_by_author.png')
```

# Dichotomy Distribution

Four dichotomies:
- introversion vs extroversion (I vs E)
- sensing vs intuition (S vs N)
- thinking vs. feeling (T vs F)
- perception vs judging (P vs J)

```python
dichotomy = twisty_sample.groupby(['mbti', 'language']).count()['user_id'].reset_index()
dichotomy
```

```python
I = dichotomy[dichotomy['mbti'].apply(lambda x: 'I' in x)]
E = dichotomy[~dichotomy['mbti'].apply(lambda x: 'I' in x)]
print("Percent Introvert:", I['user_id'].sum()/twisty_sample.shape[0]*100)
print("Percent Extrovert:", 100-I['user_id'].sum()/twisty_sample.shape[0]*100)
```

```python
S = dichotomy[dichotomy['mbti'].apply(lambda x: 'S' in x)]
N = dichotomy[~dichotomy['mbti'].apply(lambda x: 'S' in x)]
print("Percent Sensing:", S['user_id'].sum()/twisty_sample.shape[0]*100)
print("Percent Intuition:", 100-S['user_id'].sum()/twisty_sample.shape[0]*100)
```

```python
T = dichotomy[dichotomy['mbti'].apply(lambda x: 'T' in x)]
F = dichotomy[~dichotomy['mbti'].apply(lambda x: 'T' in x)]
print("Percent Thinking:", T['user_id'].sum()/twisty_sample.shape[0]*100)
print("Percent Feeling:", 100-T['user_id'].sum()/twisty_sample.shape[0]*100)
```

```python
P = dichotomy[dichotomy['mbti'].apply(lambda x: 'P' in x)]
J = dichotomy[~dichotomy['mbti'].apply(lambda x: 'P' in x)]
print("Percent Thinking:", P['user_id'].sum()/twisty_sample.shape[0]*100)
print("Percent Feeling:", 100-P['user_id'].sum()/twisty_sample.shape[0]*100)
```
